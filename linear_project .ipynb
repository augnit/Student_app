```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#label encoder and standerdscaling
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le = LabelEncoder()
```

```python
placement["Extracurricular Activities"] = le.fit_transform(placement['Extracurricular Activities'])
```

```python
placement
```

```python
X = placement.drop("Performance Index",axis=1)
```

```python
Y = placement.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 5)
```

```python
X_train
```

```python
scaler = StandardScaler()
```

```python
x_train_data = scaler.fit_transform(X_train)
x_test_data = scaler.fit_transform(X_test)
```

```python
x_train_data
```

```python
from sklearn.linear_model import LinearRegression
import pickle
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
linear = LinearRegression()
```

```python
linear.fit(x_train_data,Y_train)
```

```python
y_pred = linear.predict(x_test_data)
```

```python
x_test_data
```

```python

mean_squared_error(Y_test,y_pred)
```

```python
r2_score(Y_test,y_pred)
```

```python
# saving my model in local desktop as file 
import pickle
with open("Studenet_final_lr_model.pkl","wb") as file:
    pickle.dump((linear,scaler,le),file)
```
